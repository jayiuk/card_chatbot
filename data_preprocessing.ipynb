```python
import json
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
path = "./Training/labeled"
files = os.listdir(path)
files
```

```python
data_time = pd.DataFrame()
for file in files:
    with open(os.path.join(path, file), 'r') as f:
        js = json.load(f)
    data_list = []
    for utterance in js["utterances"]:
        data_list.append({utterance['utterance_id'] : utterance['text']})
    df = pd.DataFrame(data_list)
    data_time = pd.concat([data_time, df])
```
